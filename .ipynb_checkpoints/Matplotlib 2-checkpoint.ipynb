```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_stu_age = np.random.randint(5,40,100)
py_stu_age = np.random.randint(10,45,100)
print(ml_stu)
print(py_stu)
```

```python
plt.hist(ml_stu_age)
plt.title("Machine Learning Students")
plt.xlabel("Student's Age")
plt.ylabel("No. of students")
# plt.figure(figsize=10)
plt.show()
```

```python
plt.hist([ml_stu_age, py_stu_age], bins = [5,10,15,20,25,30,35,40,45],rwidth=0.8, color=['g','b'])
# plt.figure(figsize=(16,10))
plt.title("Machine Learning and Python Students")
plt.xlabel("Student's Age")
plt.ylabel("No. of students")
plt.show()
```

```python
plt.hist([ml_stu_age, py_stu_age], bins = [5,10,15,20,25,30,35,40,45],rwidth=0.8, color=['g','b'], orientation='horizontal')
plt.figure(figsize=(16,10))
plt.title("Machine Learning and Python Students")
plt.xlabel("Student's Age")
plt.ylabel("No. of students")
plt.show()
```

```python
plt.figure(figsize=(16,7)) # ye hist function se phele lgana hai
plt.hist([ml_stu_age, py_stu_age], bins = [5,10,15,20,25,30,35,40,45],rwidth=0.8, color=['g','b'], label=['ML stu', 'Py stu'])
plt.title("Machine Learning and Python Students", fontsize=20)
plt.xlabel("Student's Age", fontsize=13)
plt.ylabel("No. of students", fontsize=13)
plt.legend(shadow=True, fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(16,7)) # ye hist function se phele lgana hai
ax = plt.gca()  # Get the current axes

# Change the background color of the histogram area
ax.set_facecolor('black')  # Replace 'lightblue' with your desired color
plt.hist([ml_stu_age, py_stu_age], bins = [5,10,15,20,25,30,35,40,45],rwidth=0.8, color=['r','y'], label=['ML stu', 'Py stu'])
plt.title("Machine Learning and Python Students", fontsize=20)
plt.xlabel("Student's Age", fontsize=13)
plt.ylabel("No. of students", fontsize=13)
plt.legend(shadow=True, fontsize=13)
plt.show()
```

```python
plt.hist([ml_stu_age, py_stu_age], bins = [5,10,15,20,25,30,35,40,45],rwidth=0.8, color=['r','y'], label=['ML stu', 'Py stu'],range=(10,40))
plt.title("Machine Learning and Python Students", fontsize=20)
plt.xlabel("Student's Age", fontsize=13)
plt.ylabel("No. of students", fontsize=13)
plt.legend(shadow=True)
plt.show()
```

```python
plt.hist([ml_stu_age, py_stu_age],rwidth=0.8, color=['r','y'], label=['ML stu', 'Py stu'],range=(10,40)) #range lgane se outliers remove hogae jha tk ka data chaeye wo just mention krdo 
plt.title("Machine Learning and Python Students", fontsize=20)
plt.xlabel("Student's Age", fontsize=13)
plt.ylabel("No. of students", fontsize=13)
plt.legend(shadow=True)
plt.show()
```

```python
plt.hist([ml_stu_age, py_stu_age],rwidth=0.8, color=['r','y'], label=['ML stu', 'Py stu'],stacked=True) #stacked ek dataset ko dosre per rakh det hai
plt.xlabel("Student's Age", fontsize=13)
plt.ylabel("No. of students", fontsize=13)
plt.legend(shadow=True)
plt.show()
```

```python
plt.hist([ml_stu_age, py_stu_age],rwidth=0.8, color=['r','y'], label=['ML stu', 'Py stu'], histtype='step') 
plt.xlabel("Student's Age", fontsize=13)
plt.ylabel("No. of students", fontsize=13)
plt.legend(shadow=True)
plt.show()
```

```python
plt.hist([ml_stu_age, py_stu_age],rwidth=0.8, color=['r','y'], label=['ML stu', 'Py stu'], cumulative=True) 
plt.xlabel("Student's Age", fontsize=13)
plt.ylabel("No. of students", fontsize=13)
plt.legend(shadow=True)
plt.show()
```

```python
plt.hist([ml_stu_age, py_stu_age],rwidth=0.8, color=['r','y'], label=['ML stu', 'Py stu'], cumulative= -1) 
plt.xlabel("Student's Age", fontsize=13)
plt.ylabel("No. of students", fontsize=13)
plt.legend(shadow=True)
plt.show()
```

```python
plt.hist([ml_stu_age, py_stu_age],rwidth=0.8, color=['r','y'], label=['ML stu', 'Py stu'], density=True) 
plt.xlabel("Student's Age", fontsize=13)
plt.ylabel("No. of students", fontsize=13)
plt.legend(shadow=True)
plt.show()
```

How It Works: When you set density=True, the height of each bar in the histogram represents how common the data is relative to the total number of data points. The total area under all the bars equals 1, which makes it easier to compare different datasets or understand the distribution of your data.

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%',colors=['c','b','r','y','g'], center=(2,3))
plt.legend(loc = 'center left', fontsize='small')

plt.show()
```

```python
wedgeprops = {
    'edgecolor': 'black',
    'linewidth': 2,
    'width' : 3
}
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%',colors=['c','b','r','y','g'], wedgeprops=wedgeprops)
plt.legend(loc = 'center left', fontsize='small')

plt.show()
```
